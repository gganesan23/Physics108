```python
import time
import numpy as np
import matplotlib.pyplot as plt
from IPython import display

%run k2400.ipynb
%run srs.ipynb
%run fileio.ipynb


def lockin_iv_testing(lockin1,lockin2,lockin3,delay):
    %matplotlib inline
    
    
    end_time = 100000
    times = np.linspace(0,end_time,round(end_time/delay))

    lockin1vals = np.zeros(len(times))
    lockin2vals = np.zeros(len(times))
    lockin3vals = np.zeros(len(times))
    lockin3auxvals = np.zeros(len(times))
    lockin3phase = np.zeros(len(times))    

    # define a figure with two panels to show the data being acquired
    fig = plt.figure(figsize=(8, 3)) 
    
    fig.suptitle('measurements')
    
    ax1=plt.subplot(2,2,1)
    ax1.set_xlabel('time')
    ax1.set_ylabel('lockin 1 in')
    plt.plot(times,lockin1vals)
    
    ax2=plt.subplot(2,2,2)
    ax1.set_xlabel('time')
    ax2.set_ylabel('lockin 2 in')
    plt.plot(times,lockin2vals)
    
    ax3=plt.subplot(2,2,3)
    ax1.set_xlabel('time')
    ax1.set_ylabel('lockin 3 in')
    plt.plot(times,lockin3vals)
    
    ax4=plt.subplot(2,2,4)
    ax1.set_xlabel('time')
    ax4.set_ylabel('lockin 3 aux in')
    plt.plot(times,lockin3auxvals)
    
    plt.tight_layout()
    
    
    try:
        # the for loop below is the main acquisition loop. the reason the for loop lives in a try-except
        # is that if the user cancels while the sweep is running, which throws an exception, the program 
        # catches it and continues to execute below the except statement. this lets the program shut down
        # nicely and save what data has already been collected.
        
        # main loop: iterate over dc voltages
        for i in range(len(times)):
            
            # wait acquisition delay             
            lockin1vals[i] = readsrs(lockin1)[0]
            lockin2vals[i] = readsrs(lockin2)[0]
            lockin3vals[i], lockin3phase[i], lockin3auxvals[i] = readsrs(lockin3)
            print('read')
                        
            # update plots
            ax1.lines[0].set_xdata(times[0:(i+1)])
            ax1.lines[0].set_ydata(lockin1vals[0:(i+1)])
            ax1.relim()
            ax1.autoscale()

            ax2.lines[0].set_xdata(times[0:(i+1)])
            ax2.lines[0].set_ydata(lockin2vals[0:(i+1)])
            ax2.relim()
            ax2.autoscale()
            
            ax3.lines[0].set_xdata(times[0:(i+1)])
            ax3.lines[0].set_ydata(lockin3vals[0:(i+1)])
            ax3.relim()
            ax3.autoscale()

            ax4.lines[0].set_xdata(times[0:(i+1)])
            ax4.lines[0].set_ydata(lockin3auxvals[0:(i+1)])
            ax4.relim()
            ax4.autoscale()
            
            plt.tight_layout()

            display.clear_output(wait=True)
            display.display(plt.gcf())
            time.sleep(delay)
            print('got to ' + str(i*3) + ' seconds')
            print(lockin1vals[i] == lockin2vals[i] == lockin3vals[i] == lockin3auxvals[i])

    except: 
        return [lockin1vals, lockin2vals, lockin3vals, lockin3auxvals]
         # do nothing if there was an exception. just proceed to what follows the try-except statement
        
    display.clear_output(wait=True)
    
    return [lockin1vals, lockin2vals, lockin3vals, lockin3auxvals]
    
    # save data as a .pkl file for use with python
    
#     # also save data as a text file for easy opening with other software
#     np.savetxt(full_filename+'.txt',np.transpose(np.array([iv.dc_voltage,iv.dc_current,iv.ac_voltagexx,iv.phase,iv.temp])),newline='\r\n')
#     np.savetxt(full_filename2+'.txt',np.transpose(np.array([iv2.dc_voltage,iv2.dc_current,iv2.ac_voltagexx,iv2.phase,iv2.temp])),newline='\r\n')
```
